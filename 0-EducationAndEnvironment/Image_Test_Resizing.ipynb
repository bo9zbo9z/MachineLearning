```python
# Normal includes...
from __future__ import absolute_import, division, print_function, unicode_literals

import os, sys, random, warnings, time, copy, csv
import numpy as np 

import IPython.display as display
from PIL import Image

import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
print(tf.__version__)
```

```python
# Loads image from file path, assums jpeg, change decode if png
def image_load(file_path):
    image = tf.io.read_file(file_path)
    image = tf.image.decode_jpeg(image, channels=3)  # CHANGE BASED ON IMAGE FILE TYPE
    image = tf.image.convert_image_dtype(image, tf.float32)
    return image
```

```python
# Location and resizing of your test image, change as needed...

FILE_PATH = "/Users/john/Documents/Python-Working/muppets.jpeg"  ###### CHANGE FOR SPECIFIC ENVIRONMENT
IMG_DIM = (128, 128)
PLT_DIM = (20,20)
```

```python
# https://www.tensorflow.org/api_docs/python/tf/image/resize

plt.figure(figsize = PLT_DIM)
plt.imshow(image_load(FILE_PATH))
plt.title("Original Image")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM))
plt.title("Original Image Default Resized")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM, method=tf.image.ResizeMethod.BILINEAR))
plt.title("BILINEAR")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM, method=tf.image.ResizeMethod.BICUBIC))
plt.title("BICUBIC")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM, method=tf.image.ResizeMethod.NEAREST_NEIGHBOR))
plt.title("NEAREST_NEIGHBOR")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM, method=tf.image.ResizeMethod.AREA))
plt.title("AREA")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM, method=tf.image.ResizeMethod.LANCZOS3))
plt.title("LANCZOS3")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM, method=tf.image.ResizeMethod.LANCZOS5))
plt.title("LANCZOS5")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM, method=tf.image.ResizeMethod.GAUSSIAN))
plt.title("GAUSSIAN")
plt.show()

plt.figure(figsize = PLT_DIM)
plt.imshow(tf.image.resize(image_load(FILE_PATH), size=IMG_DIM, method=tf.image.ResizeMethod.MITCHELLCUBIC))
plt.title("MITCHELLCUBIC")
plt.show()
```
